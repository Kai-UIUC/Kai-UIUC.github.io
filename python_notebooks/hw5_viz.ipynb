```python
import altair as alt

url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv"

alt.data_transformers.disable_max_rows()

# Plot 1: Bar chart with tooltips
plot1 = alt.Chart(url).mark_bar().encode(
    x=alt.X('season:N', title='Season', sort=['Spring', 'Summer', 'Fall', 'Winter']),
    y=alt.Y('count():Q', title='Number of Reports'),
    color=alt.Color('classification:N', title='Classification'),
    tooltip=['season:N', 'classification:N', alt.Tooltip('count():Q', title='Count')]
).properties(
    title='Bigfoot Reports by Season and Classification',
    width=400,
    height=300
)

plot1.display()

# Plot 2: Scatter plot with dropdown filter for season
season_dropdown = alt.binding_select(
    options=[None, 'Spring', 'Summer', 'Fall', 'Winter'],
    labels=['All Seasons', 'Spring', 'Summer', 'Fall', 'Winter'],
    name='Select Season: '
)

season_select = alt.selection_point(fields=['season'], bind=season_dropdown)

plot2 = alt.Chart(url).mark_circle(size=50, opacity=0.6).encode(
    x=alt.X('temperature_mid:Q', title='Average Temperature'),
    y=alt.Y('humidity:Q', title='Humidity'),
    color=alt.condition(season_select,
                        alt.Color('classification:N', title='Classification'),
                        alt.value('lightgray')),
    tooltip=['state:N', 'county:N', 'temperature_mid:Q', 'humidity:Q', 'season:N']
).add_params(
    season_select
).properties(
    title='Weather Conditions During Bigfoot Sightings',
    width=500,
    height=400
)

plot2.display()

# Save charts to JSON files for the Jekyll website
plot1.save('bigfoot_plot1.json')
plot2.save('bigfoot_plot2.json')
```
